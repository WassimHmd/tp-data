```python
import geopandas as gpd
import pandas as pd
from geopandas import GeoDataFrame, GeoSeries
import matplotlib.pyplot as plt

gdf = gpd.read_file("data\landcover\dz\dza_gc_adg.shp")

```

```python
print(gdf.head())
gdf.plot(column="GRIDCODE", legend=True)
```

```python
gdf_tn = gpd.read_file("data\landcover\\tn\\tun_gc_adg.shp")
```

```python
print(gdf_tn.head())
gdf_tn.plot()
```

```python
# Reproject if needed
# tun = tun.to_crs(alg.crs)

# Combine
combined = gpd.GeoDataFrame(pd.concat([gdf, gdf_tn], ignore_index=True))
```

```python

# Plot merged shapefile
combined.plot(column="GRIDCODE", legend=True)
plt.title("Merged Algeria + Tunisia")
plt.show()
```
